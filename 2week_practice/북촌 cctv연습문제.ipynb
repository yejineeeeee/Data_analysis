## JSON 형식의 데이터 이해하기

### 1. 연습문제 1번에서 문자열로 저장한 data 변수의 데이터를 파이썬 딕셔너리로 변환하세요. 변환된 결과를 dict_data 변수에 저장하세요.

```python
# json.loads() 함수는 JSON 형태의 문자열을 입력으로 받아들이고 Python의 딕셔너리 자료형으로 변환합니다.
#이렇게 문자열을 객체로 변환하는 과정을 역직렬화(deserialization) 라고 합니다.
import json
dict_data = json.loads(data) # 딕셔너리로 바꾸기
print(dict_data)
```

```python
print(type(dict_data))
```

### 3. dict_data 변수에 저장된 딕셔너리의 key 배열을 출력하세요.

```python
dict_data.keys()

dict_data['BukChonInOutPeopleInfo'].keys()
```

### 4. dict_data 변수에 저장된 딕셔너리의 ‘BukChonInOutPeopleInfo' key에 해당하는 데이터는 3개의 key가 있는 딕셔너리입니다. 여기서 ‘row’ key에는 5개의 딕셔너리가 원소인 리스트가 들어 있습니다. ‘row’ key의 리스트 데이터를 판다스 데이터프레임으로 변환하고 변수 df에 저장하세요.

```python
df = pd.DataFrame(dict_data['BukChonInOutPeopleInfo']['row'])
df
```

## 연습문제3. 월별 데이터를 집계하기
#### 불러온 1000개의 데이터 중 1월에 해당하는 데이터가 적어 11월과 12월의 데이터로 문제를 풀었습니다.

### 1. df 데이터프레임의 ‘ENDTIME’ 변수에서 월(MONTH)을 추출하고 ‘MONTH’ 열을 추가하여 저장하세요. 

```python
df['ENDTIME'] = pd.to_datetime(df['ENDTIME'])

df['MONTH'] = df['ENDTIME'].dt.month
df
```

### 2. 11월 데이터의 ‘INCOUNT’ 데이터의 평균값을 구해서 출력하세요. 

```python
df_11 = df[df['MONTH'] == 11]
df_11['INCOUNT'].mean()
```

### 3. 10월과 11월의 ‘OUTCOUNT’ 데이터의 평균값을 구해서 출력하세요. 

```python
df_10 = df[df['MONTH'] == 10]

df_10_11 = pd.concat([df_10,df_11])
df_10_11.groupby(['MONTH'])['OUTCOUNT'].mean()
```

### 4. 10월과 11월의 ‘INCOUNT’ 와 ‘OUTCOUNT’ 데이터의 최솟값과 최댓값을 구하세요. 

```python
df_10_11.groupby(['MONTH'])['INCOUNT', 'OUTCOUNT'].agg(['min', 'max'])
```

### 5. 10월과 11월의 ‘OUTCOUNT’ 데이터의 평균값을 seaborn 라이브러리의 barplot 함수를 사용하여 시각화 하세요. 

```python
import seaborn as sns

sns.barplot(data = df_10_11, x = 'MONTH', y = 'OUTCOUNT')
```
